### Polynomial Regression for Predicting Car Price 

```python
import numpy as np # linear algebra
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import ElasticNetCV
from sklearn.pipeline import make_pipeline
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import StandardScaler, PolynomialFeatures, OneHotEncoder
from sklearn.linear_model import LinearRegression

from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer


import joblib # To save our model
from flask import Flask, jsonify, request
```

```python
#Load dataset
df = pd.read_csv("CarPrice_Assignment.csv")
df.head(20)
```

```python
df.shape
```

```python
df = df.drop('car_ID',axis=1) #We drop the ID initial column

df['CarBrand'] = df['CarName'].apply(lambda x: x.split(' ')[0]) #We get the car brands 

df = df.drop('CarName',axis=1) #We drop the car name column
```

```python
df.head()
```

```python
df.info()
```

We can see that we don't have missing values

```python
# Check for duplicated rows
len(df.loc[df.duplicated()])
```

We don't have duplicated rows.

```python
# descriptive statistic
df.describe()
```

Let's split our dataset into training and test sets

```python
# Shuffle the data 
data = df.sample(frac=1,random_state=42).reset_index(drop=True)

# Calculate the number of samples for each set
train_size = int(0.8 * len(data)) 
test_size = int(0.2 * len(data))

# Split the dataset into training and test sets
train_data = data[:train_size] # training data (80%)
test_data = data[train_size:] # test data (20%)
```

```python
data.head()
```

Our target is `price`

```python
numerical_features = ['wheelbase', 'carlength', 'carwidth', 'carheight',
                      'curbweight', 'enginesize', 'boreratio', 'stroke',
                      'compressionratio', 'horsepower', 'peakrpm', 'citympg', 'highwaympg']
categorical_cols = ['CarBrand','fueltype','aspiration','doornumber','carbody','drivewheel',
                    'enginelocation','enginetype','cylindernumber','fuelsystem']

target = 'price'
```

```python
corr_matrix = df[['wheelbase', 'carlength', 'carwidth', 'carheight',
                      'curbweight', 'enginesize', 'boreratio', 'stroke',
                      'compressionratio', 'horsepower', 'peakrpm', 'citympg', 'highwaympg','price']].corr()
plt.figure(figsize=(12, 8))
sns.heatmap(
    corr_matrix,
    annot=True,
    fmt=".2f",
    cmap='coolwarm',
)
plt.title("Correlation matrix")
plt.show()
```

Some of our variables are very high correlated

```python
# Split the features and target for each set
X_train = train_data[categorical_cols + numerical_features]
y_train = train_data[target]

X_test = test_data[categorical_cols + numerical_features]
y_test = test_data[target]
```

```python

data_preprocessed = ColumnTransformer([
    ('cat',OneHotEncoder(drop='first', sparse_output=False, handle_unknown='ignore'), categorical_cols),
    ('num', 'passthrough', numerical_features)
])
```

```python
print(X_train.shape)
print(X_test.shape)
```

### Polynomial regression using `ElasticNetCV`

```python
#We create a complete pipeline
model = make_pipeline(
    data_preprocessed,
    PolynomialFeatures(degree=2, include_bias=False),
    StandardScaler(),
    ElasticNetCV(cv=5, l1_ratio=[.1,.5,.7,.9,.99,1], alphas=np.logspace(-4,2,50),max_iter=2000,tol=1e-4, random_state=42)
)

#Training
model.fit(X_train, y_train)

#Prediction
y_pred = model.predict(X_test)
```

```python
# Model evaluation
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print("\nElasticNet Polynomial Regression:")
print(f"   MSE  : {mse:.4f}")
print(f"   RMSE : {rmse:.4f}")
print(f"   MAE  : {mae:.4f}")
print(f"   R²   : {r2:.4f}")
```

```python
elastic_net = model.named_steps['elasticnetcv']
print("Best alpha : ",elastic_net.alpha_)
print("Best l1_ratio : ",elastic_net.l1_ratio_) 
```

best l1_ratio is almost equal to 1 meaning that we are using the Lasso regularization to select the most important variables.

### Let's save our model using `joblib`

```python
#save the model
joblib.dump(model, 'elastic_poly_model.joblib')
```

```python
Poly = joblib.load('elastic_poly_model.joblib')
```
